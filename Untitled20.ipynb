```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical

# --- Configuration ---
FILE_NAMES = [
    "guru_training_dataset.csv",
    "harsha_training_dataset.csv",
    "keerthana_training_dataset.csv"
]
WINDOW_SIZE = 200  # 200ms of data per slice
STEP_SIZE = 50     # 50ms slide

windows = []
labels = []

print("Loading, Normalizing, and Windowing Datasets...")

# Process each person's file individually
for file in FILE_NAMES:
    print(f"\nProcessing {file}...")
    df = pd.read_csv(file)

    # 1. User-Specific Normalization
    # We scale each person's data so the mean is 0 and standard dev is 1.
    # This removes the biological differences in baseline skin impedance!
    scaler = StandardScaler()
    df['Normalized_Voltage'] = scaler.fit_transform(df[['Voltage']])

    # 2. Windowing (Slicing the continuous stream)
    extracted_count = 0
    for i in range(0, len(df) - WINDOW_SIZE, STEP_SIZE):
        window_data = df['Normalized_Voltage'].iloc[i:i + WINDOW_SIZE].values
        window_label = df['Gesture_Label'].iloc[i:i + WINDOW_SIZE].mode()[0]

        # Ignore transition windows, setup phases, or baseline rests
        if window_label not in ["Waiting...", "Done", "Rest (Baseline)"]:
            windows.append(window_data)
            labels.append(window_label)
            extracted_count += 1

    print(f"-> Extracted {extracted_count} clean windows.")

# --- Prepare Data for Neural Network ---
X = np.array(windows)
# Reshape for Conv1D: (Samples, TimeSteps, Channels)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Encode text labels (Fist, Pinch, etc.) to One-Hot numbers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(labels)
y_categorical = to_categorical(y_encoded)

print(f"\n=== Overall Dataset ===")
print(f"Total Combined Windows: {X.shape[0]}")
print(f"Classes found to train on: {encoder.classes_}")

# Split into Training and Validation sets (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)

# --- Build the Generalized 1D CNN Model ---
print("\nBuilding Deep Learning Model...")
model = Sequential([
    # First Convolutional Block
    Conv1D(filters=32, kernel_size=5, activation='relu', input_shape=(WINDOW_SIZE, 1)),
    BatchNormalization(), # Helps stabilize training across different users
    MaxPooling1D(pool_size=2),

    # Second Convolutional Block
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    # Classifier Head
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5), # Crucial for multi-user: prevents overfitting to Guru or Harsha
    Dense(len(encoder.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# --- Train the Model ---
print("\nInitiating Training...")
history = model.fit(
    X_train, y_train,
    epochs=20,          # Bumped up slightly since we have more data
    batch_size=64,      # Larger batch size for smoother gradient updates
    validation_data=(X_test, y_test)
)

# --- Save Artifacts ---
model.save("nadicare_generalized_model.h5")
np.save("gesture_classes.npy", encoder.classes_)
print("\n✅ Multi-user model trained and saved successfully!")
print("Download 'nadicare_generalized_model.h5' and 'gesture_classes.npy' to use in your live demo.")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

# --- Configuration ---
FILE_NAMES = [
    "guru_training_dataset.csv",
    "harsha_training_dataset.csv",
    "keerthana_training_dataset.csv"
]
WINDOW_SIZE = 300  # INCREASED: Gives a better view of waveform frequency
STEP_SIZE = 50

# --- 5x Feature Extraction Functions ---
def calc_mav(window): return np.mean(np.abs(window))
def calc_wl(window): return np.sum(np.abs(np.diff(window)))
def calc_zc(window):
    centered = window - np.mean(window)
    return np.sum(np.diff(np.sign(centered)) != 0)
def calc_ssc(window): return np.sum(np.diff(np.sign(np.diff(window))) != 0)
def calc_rms(window): return np.sqrt(np.mean(window**2)) # NEW: Power feature!

features_list = []
labels = []

print("Extracting 5 Time-Domain Features from all datasets...")

for file in FILE_NAMES:
    print(f"\nProcessing {file}...")
    df = pd.read_csv(file)

    # --- THE HACKATHON SURVIVAL FILTER ---
    # We strip out Pinch and Thumbs Up to guarantee high accuracy on a single channel
    allowed_labels = ['Rest (Baseline)', 'Rest', 'Fist', 'Open Hand', 'Waiting...', 'Done']
    df = df[df['Gesture_Label'].isin(allowed_labels)]
    # -------------------------------------

    # 1. User-Specific Normalization
    scaler = StandardScaler()
    df['Normalized_Voltage'] = scaler.fit_transform(df[['Voltage']])

    extracted_count = 0
    for i in range(0, len(df) - WINDOW_SIZE, STEP_SIZE):
        window_data = df['Normalized_Voltage'].iloc[i:i + WINDOW_SIZE].values
        window_label = df['Gesture_Label'].iloc[i:i + WINDOW_SIZE].mode()[0]

        if window_label not in ["Waiting...", "Done", "Rest (Baseline)"]:
            # Extract 5 features
            mav = calc_mav(window_data)
            wl = calc_wl(window_data)
            zc = calc_zc(window_data)
            ssc = calc_ssc(window_data)
            rms = calc_rms(window_data)

            features_list.append([mav, wl, zc, ssc, rms])
            labels.append(window_label)
            extracted_count += 1

    print(f"-> Extracted {extracted_count} feature vectors.")

# --- Prepare Data ---
X = np.array(features_list)
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(labels)

print(f"\n=== Overall Dataset ===")
print(f"Total Combined Windows: {X.shape[0]}")
print(f"Classes found: {encoder.classes_}")

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# --- Train Upgraded Random Forest ---
print("\nTraining Balanced Random Forest Classifier...")
rf_model = RandomForestClassifier(
    n_estimators=200,          # INCREASED: 200 trees voting
    max_depth=None,            # REMOVED LIMIT: Let trees grow deep to find complex patterns
    class_weight='balanced',   # THE FIX: Mathematically solves your 427 vs 112 data imbalance!
    random_state=42,
    n_jobs=-1                  # Use all CPU cores for speed
)
rf_model.fit(X_train, y_train)

# --- Evaluate ---
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"\n✅ Validation Accuracy: {accuracy * 100:.2f}%")
print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))

# --- Save Artifacts ---
joblib.dump(rf_model, "nadicare_rf_model.pkl")
np.save("gesture_classes_rf.npy", encoder.classes_)
print("\n✅ Saved 'nadicare_rf_model.pkl' and 'gesture_classes_rf.npy'.")
```

```python
import numpy as np
import pandas as pd
from scipy.signal import spectrogram
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.utils.class_weight import compute_class_weight

# --- Configuration ---
FILE_NAMES = [
    "guru_training_dataset.csv",
    "harsha_training_dataset.csv",
    "keerthana_training_dataset.csv"
]
WINDOW_SIZE = 500
STEP_SIZE = 50
FS = 1000  # Sampling frequency (1000 Hz)

spectrograms = []
labels = []

print("Transforming 1D Time-Series into 2D Spectrograms...")

for file in FILE_NAMES:
    print(f"Processing {file}...")
    df = pd.read_csv(file)

    scaler = StandardScaler()
    df['Normalized_Voltage'] = scaler.fit_transform(df[['Voltage']])

    extracted_count = 0
    for i in range(0, len(df) - WINDOW_SIZE, STEP_SIZE):
        window_data = df['Normalized_Voltage'].iloc[i:i + WINDOW_SIZE].values
        window_label = df['Gesture_Label'].iloc[i:i + WINDOW_SIZE].mode()[0]

        if window_label not in ["Waiting...", "Done", "Rest (Baseline)"]:
            # THE MAGIC: Convert the 1D array into a 2D frequency heatmap
            # nperseg=64 and noverlap=48 gives us a nice dense "image" resolution
            f, t, Sxx = spectrogram(window_data, fs=FS, nperseg=64, noverlap=48)

            # Convert to Decibel scale to make the patterns "pop" visually for the CNN
            Sxx_db = 10 * np.log10(Sxx + 1e-10)

            spectrograms.append(Sxx_db)
            labels.append(window_label)
            extracted_count += 1

    print(f"-> Extracted {extracted_count} spectrogram images.")

# --- Prepare Data for Computer Vision ---
X = np.array(spectrograms)
# Reshape for Conv2D: (Samples, Height/Freqs, Width/Time, Channels)
# Shape will be approximately (N, 33, 16, 1)
X = X.reshape(X.shape[0], X.shape[1], X.shape[2], 1)

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(labels)
y_categorical = to_categorical(y_encoded)

print(f"\n=== Overall Dataset ===")
print(f"Spectrogram Image Shape: {X.shape[1]}x{X.shape[2]} pixels")
print(f"Total Combined Windows: {X.shape[0]}")
print(f"Classes found: {encoder.classes_}")

X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)

# Calculate class weights to handle the "Rest" imbalance in Deep Learning
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_encoded),
    y=y_encoded
)
class_weight_dict = dict(enumerate(class_weights))

# --- Build the 2D Spectrogram CNN ---
print("\nBuilding 2D Deep Learning Model...")
model = Sequential([
    # Look for local frequency bursts
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(X.shape[1], X.shape[2], 1)),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    # Look for complex patterns across time and frequency
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    BatchNormalization(),
    # MaxPooling2D(pool_size=(2, 2)),

    # Look for complex patterns across time and frequency
    Conv2D(128, kernel_size=(3, 3), activation='relu'),
    BatchNormalization(),
    # MaxPooling2D(pool_size=(2, 2)),

    # Look for complex patterns across time and frequency
    Conv2D(256, kernel_size=(3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.6), # Heavy dropout to prevent overfitting to specific team members
    Dense(len(encoder.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# --- Train ---
print("\nInitiating Training...")
history = model.fit(
    X_train, y_train,
    epochs=30, # Needs a bit more time to learn images
    batch_size=64,
    class_weight=class_weight_dict, # Pass our balancing weights here
    validation_data=(X_test, y_test)
)

model.save("nadicare_spectrogram_model.h5")
np.save("gesture_classes_spec.npy", encoder.classes_)
print("\n✅ Spectrogram model trained and saved successfully!")
```

```python
import numpy as np
import pandas as pd
import joblib
from scipy.signal import iirnotch, butter, filtfilt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# --- Configuration ---
# List ALL 12 of your files here
FILE_NAMES = [
    "guru1.csv", "guru2.csv", "guru3.csv", "guru4.csv",
    "harsha1.csv", "harsha2.csv", "harsha3.csv", "harsha4.csv",
    "harshitha1.csv", "harshitha2.csv", "harshitha3.csv", "harshitha4.csv"
]

WINDOW_SIZE = 200
STEP_SIZE = 50
SAMPLE_RATE = 1000

# --- Advanced Signal Processing ---
def apply_filters(data):
    # 1. Digital DRL (50Hz Notch Filter) to fix Guru's data
    notch_freq = 50.0
    quality_factor = 30.0
    b_notch, a_notch = iirnotch(notch_freq, quality_factor, SAMPLE_RATE)
    data_notched = filtfilt(b_notch, a_notch, data)

    # 2. Bandpass Filter (20Hz - 450Hz) to remove movement sway
    nyquist = SAMPLE_RATE / 2.0
    low = 20.0 / nyquist
    high = 450.0 / nyquist
    b_band, a_band = butter(4, [low, high], btype='band')
    clean_data = filtfilt(b_band, a_band, data_notched)

    return clean_data

# --- Robust Feature Extraction ---
def extract_features(window):
    mav = np.mean(np.abs(window))
    rms = np.sqrt(np.mean(window**2))
    var = np.var(window)
    wl = np.sum(np.abs(np.diff(window)))
    zc = np.sum(np.diff(np.sign(window)) != 0)

    # NEW: Slope Sign Changes (Highly robust to sensor sliding)
    diffs = np.diff(window)
    ssc = np.sum(np.diff(np.sign(diffs)) != 0)

    return [mav, rms, var, wl, zc, ssc]

X_features = []
labels = []

print("Initiating Nadicare Master Pipeline...")

for file in FILE_NAMES:
    print(f"\nProcessing {file}...")
    df = pd.read_csv(file)

    allowed_labels = ['Rest (Baseline)', 'Rest', 'Fist', 'Open Hand']
    df = df[df['Gesture_Label'].isin(allowed_labels)].copy()

    # --- CRITICAL: Apply filters to the CONTINUOUS data before windowing ---
    print("   -> Applying Digital DRL and Bandpass filters...")
    df['Filtered_Voltage'] = apply_filters(df['Voltage'].values)

    # Normalize the clean data
    scaler = StandardScaler()
    df['Normalized_Voltage'] = scaler.fit_transform(df[['Filtered_Voltage']])

    extracted_count = 0
    # Window the data
    for i in range(0, len(df) - WINDOW_SIZE, STEP_SIZE):
        window_data = df['Normalized_Voltage'].iloc[i:i + WINDOW_SIZE].values
        window_label = df['Gesture_Label'].iloc[i:i + WINDOW_SIZE].mode()[0]

        if window_label == 'Rest (Baseline)':
            window_label = 'Rest'

        features = extract_features(window_data)
        X_features.append(features)
        labels.append(window_label)
        extracted_count += 1

    print(f"   -> Extracted {extracted_count} robust feature vectors.")

X = np.array(X_features)
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(labels)

print(f"\n=== Final Combined Dataset ===")
print(f"Total Windows: {X.shape[0]}")
print(f"Classes: {encoder.classes_}")

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# --- Optimized Random Forest ---
print("\nTraining Optimized Random Forest...")
# Increased estimators and adjusted max_depth to handle the massive 1.5M row variance
rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=20,
    min_samples_leaf=2,
    class_weight='balanced',
    n_jobs=-1, # Uses all CPU cores in Colab for faster training
    random_state=42
)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\n🚀 FINAL VALIDATION ACCURACY: {acc*100:.2f}%\n")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))

joblib.dump(rf_model, "nadicare_robust_rf.pkl")
np.save("nadicare_classes.npy", encoder.classes_)
print("✅ Models saved! Ready for live testing.")
```

```python
import numpy as np
import pandas as pd
import joblib
from scipy.signal import iirnotch, butter, filtfilt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from xgboost import XGBClassifier

# --- Configuration ---
FILE_NAMES = [
    "guru1.csv", "guru2.csv", "guru3.csv", "guru4.csv",
    "harsha1.csv", "harsha2.csv", "harsha3.csv", "harsha4.csv",
    "harshitha1.csv", "harshitha2.csv", "harshitha3.csv", "harshitha4.csv"
]

WINDOW_SIZE = 200
STEP_SIZE = 50
SAMPLE_RATE = 1000

# --- Advanced Signal Processing ---
def apply_filters(data):
    # Digital DRL (50Hz Notch)
    b_notch, a_notch = iirnotch(50.0, 30.0, SAMPLE_RATE)
    data_notched = filtfilt(b_notch, a_notch, data)

    # Bandpass (20Hz - 450Hz)
    nyquist = SAMPLE_RATE / 2.0
    b_band, a_band = butter(4, [20.0 / nyquist, 450.0 / nyquist], btype='band')
    clean_data = filtfilt(b_band, a_band, data_notched)

    return clean_data

# --- The "Heavy Artillery" Feature Extractor ---
def extract_features(window):
    # 1. TIME DOMAIN
    mav = np.mean(np.abs(window))
    rms = np.sqrt(np.mean(window**2))
    var = np.var(window)
    wl = np.sum(np.abs(np.diff(window)))
    zc = np.sum(np.diff(np.sign(window)) != 0)
    ssc = np.sum(np.diff(np.sign(np.diff(window))) != 0)

    # 2. HJORTH PARAMETERS (Signal Texture)
    activity = var
    diff1 = np.diff(window)
    mobility = np.sqrt(np.var(diff1) / activity) if activity > 0 else 0
    diff2 = np.diff(diff1)
    complexity = np.sqrt(np.var(diff2) / np.var(diff1)) / mobility if mobility > 0 else 0

    # 3. FREQUENCY DOMAIN (Fast Fourier Transform)
    # FFT calculates the power of different frequencies in the muscle
    fft_vals = np.abs(np.fft.rfft(window))
    freqs = np.fft.rfftfreq(len(window), d=1.0/SAMPLE_RATE)

    # Total Spectral Power
    total_power = np.sum(fft_vals)
    # Mean Frequency (Center of mass of the frequency spectrum)
    mean_freq = np.sum(freqs * fft_vals) / total_power if total_power > 0 else 0

    # Peak Frequency (The dominant firing frequency)
    peak_freq = freqs[np.argmax(fft_vals)] if total_power > 0 else 0

    return [mav, rms, var, wl, zc, ssc, activity, mobility, complexity, mean_freq, peak_freq]

X_features = []
labels = []

print("🔥 Initiating Nadicare XGBoost Master Pipeline...")

for file in FILE_NAMES:
    print(f"\nProcessing {file}...")
    df = pd.read_csv(file)

    allowed_labels = ['Rest (Baseline)', 'Rest', 'Fist', 'Open Hand']
    df = df[df['Gesture_Label'].isin(allowed_labels)].copy()

    print("   -> Filtering data...")
    df['Filtered_Voltage'] = apply_filters(df['Voltage'].values)

    scaler = StandardScaler()
    df['Normalized_Voltage'] = scaler.fit_transform(df[['Filtered_Voltage']])

    extracted_count = 0
    for i in range(0, len(df) - WINDOW_SIZE, STEP_SIZE):
        window_data = df['Normalized_Voltage'].iloc[i:i + WINDOW_SIZE].values
        window_label = df['Gesture_Label'].iloc[i:i + WINDOW_SIZE].mode()[0]

        if window_label == 'Rest (Baseline)':
            window_label = 'Rest'

        features = extract_features(window_data)
        X_features.append(features)
        labels.append(window_label)
        extracted_count += 1

    print(f"   -> Extracted {extracted_count} deep feature vectors.")

X = np.array(X_features)
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(labels)

print(f"\n=== Final Combined Dataset ===")
print(f"Total Windows: {X.shape[0]}")
print(f"Classes: {encoder.classes_}")

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# --- The XGBoost Model ---
print("\n🚀 Training Extreme Gradient Boosting (XGBoost)...")
xgb_model = XGBClassifier(
    n_estimators=500,          # More trees to map the massive variance
    learning_rate=0.05,        # Slow learning for high precision
    max_depth=10,              # Deep trees to handle the non-linear sensor shifts
    subsample=0.8,             # Prevents overfitting
    colsample_bytree=0.8,      # Prevents overfitting
    objective='multi:softprob',
    num_class=len(encoder.classes_),
    eval_metric='mlogloss',
    n_jobs=-1,
    random_state=42
)

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\n🏆 FINAL VALIDATION ACCURACY: {acc*100:.2f}%\n")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))

# Print Feature Importance so you can literally see what's working
print("\n📊 Feature Importance Rankings:")
feature_names = ["MAV", "RMS", "Variance", "Waveform Length", "Zero Crossings", "SSC", "Hjorth Activity", "Hjorth Mobility", "Hjorth Complexity", "Mean Frequency", "Peak Frequency"]
importances = xgb_model.feature_importances_
for name, imp in sorted(zip(feature_names, importances), key=lambda x: x[1], reverse=True):
    print(f" - {name}: {imp:.4f}")

joblib.dump(xgb_model, "nadicare_xgboost_model.pkl")
np.save("nadicare_classes_xgb.npy", encoder.classes_)
print("✅ XGBoost Model saved! Ready to dominate.")
```
